```python
#Import Libraries
import pandas as pd
import numpy as np 
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams["figure.figsize"]=(20,10)
```

```python
#Read CSV File
df1 = pd.read_csv("Data_File.csv")
df1.head()
```

```python
#Drop Rows that have any null values
df2 = df1.dropna()
df2.head()
```

```python
#Transforming the size column to only have numerical values as bedrooms
df3=df2.copy()
df3['bedrooms'] = df3['size'].apply(lambda x: int(x.split(' ')[0]))
df3 = df3.drop(['size'],axis='columns')
df3.head()
```

```python
#Converting Rnage of Square Feet values to one average value
def convert_range_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
    
df4 = df3.copy()
df4['total_sqft'] = df4['total_sqft'].apply(convert_range_to_num)
df4.head()
```

```python
#Calculating price per sqaure feet
df5 = df4.copy()
df5['price_sqft']=(df5['price']*100000)/df5['total_sqft']
df5.head()
```

```python
#Renaming locations less than 10 to be named "Other"
df5.location = df5.location.apply(lambda x: x.strip())
LocStats = df5.groupby('location')['location'].agg('count').sort_values(ascending=False)
LocationLT10 = LocStats[LocStats<=10]
df5.location = df5.location.apply(lambda x: 'Other' if x in LocationLT10 else x)
df5.head()
```

```python
#removing outliers using domain knowledge
#calculating size per bedroom to check outliers
df5[df5.total_sqft/df5.bedrooms<300].head()
```

```python
df5.shape
```

```python
#removing size per bedroom less than 300
df6 = df5[~(df5.total_sqft/df5.bedrooms<300)]
df6.shape
```

```python
df6.price_sqft.describe()
```

```python
#removing extreme price per squareft.
#per location mean and std and then remove anything beyond 1 std.
def remove_pricepersqft_outliers(df):
    df_out = pd.DataFrame()
    for key,subdf in df.groupby('location'):
        m = np.mean(subdf.price_sqft)
        st = np.std(subdf.price_sqft)
        reduced_df = subdf[(subdf.price_sqft>(m-st)) & (subdf.price_sqft<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index = True)
    return df_out
df7 = remove_pricepersqft_outliers(df6)
df7.shape
```

```python
#drawing scatterplot for price changes depending on bedroom number
def plot_scatter_chart(df,location):
    bedrooms2 = df[(df.location == location) & (df.bedrooms == 2)]
    bedrooms3 = df[(df.location == location) & (df.bedrooms == 3)]
    matplotlib.rcParams['figure.figsize'] = (15,10)
    plt.scatter(bedrooms2.total_sqft, bedrooms2.price,color = 'blue',label = '2 bedrooms', s= 50)
    plt.scatter(bedrooms3.total_sqft, bedrooms3.price,marker = '+',color = 'green',label = '3 bedrooms', s= 50)
    plt.xlabel("Total Square Feet Area")
    plt.ylabel("Price")
    plt.title(location)
    plt.legend()

plot_scatter_chart(df7,"Electronic City")
```

```python
#removing properties where for same location, price of 3bd apart. is
#lesser than mean of 2bd apart and same square ft. area.
def remove_bedrooms_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bedrooms_stats = {}
        for bedrooms,bedrooms_df in location_df.groupby('bedrooms'):
            bedrooms_stats[bedrooms] = {
                'mean': np.mean(bedrooms_df.price_sqft),
                'std': np.std(bedrooms_df.price_sqft),
                'count': bedrooms_df.shape[0]
            }
        for bedrooms,bedrooms_df in location_df.groupby('bedrooms'):
            stats = bedrooms_stats.get(bedrooms-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices,bedrooms_df[bedrooms_df.price_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis = 'index')

df8 = remove_bedrooms_outliers(df7)
df8.shape
```

```python
#drawing scatterplot after bedroom outlier removal
def plot_scatter_chart(df,location):
    bedrooms2 = df[(df.location == location) & (df.bedrooms == 2)]
    bedrooms3 = df[(df.location == location) & (df.bedrooms == 3)]
    matplotlib.rcParams['figure.figsize'] = (15,10)
    plt.scatter(bedrooms2.total_sqft, bedrooms2.price,color = 'blue',label = '2 bedrooms', s= 50)
    plt.scatter(bedrooms3.total_sqft, bedrooms3.price,marker = '+',color = 'green',label = '3 bedrooms', s= 50)
    plt.xlabel("Total Square Feet Area")
    plt.ylabel("Price")
    plt.title(location)
    plt.legend()

plot_scatter_chart(df8,"Electronic City")
```

```python
#no.of bathroom > no.of bedrooms+2 is an outlier. remove those

df9 = df8[df8.bath<df8.bedrooms+2]
df9.shape
```

```python
#REMOVING EXCESS DATA
#only needed price per square ft. for outlier removal
#Availability is simply boolean
#Society and Area_Type are text oriented and could be judged on the basis of location
#Balcony is not that relevant, however could be used if customer wants.
df10 = df9.drop(['price_sqft','availability','society','area_type','balcony'],axis = 'columns')
df10.head()
```

```python
dummies = pd.get_dummies(df10.location)
dummies.head()
```

```python
#Dropping Others column so that all zeros mean it is 1
df11 = pd.concat([df10,dummies.drop('Other',axis='columns')],axis='columns')
df11.head()
```

```python
#No longer need location column as all is in numbers
df12 = df11.drop(['location'],axis='columns')
df12.head()
```

```python
#defining X-Axis of the graph
X_AXIS = df12.drop(['price'],axis='columns')
X_AXIS.head()
```

```python
#defining Y-Axis of the graph
Y_AXIS = df12.price
Y_AXIS.head()
```

```python
#Setting the training and testing data
from sklearn.model_selection import train_test_split
X_AXIS_train, X_AXIS_test, Y_AXIS_train, Y_AXIS_test = train_test_split(X_AXIS,Y_AXIS,test_size=0.2,random_state=10)
```

```python
#Using Linear Regression to fit the curve and predict a score
#We can also use multiple other methods(Decision Tree, Lasso etc..) but Linear Regression is giving best accuracy
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_AXIS_train,Y_AXIS_train)
lr_clf.score(X_AXIS_test,Y_AXIS_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
#Randomizes Samples so each fold has equal distribution of data
cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
#K-Fold Cross Validation
cross_val_score(LinearRegression(),X_AXIS,Y_AXIS,cv=cv)
```

```python
#Function to predict scores using Linear Regression
def PredictPrice(location,sqft,bath,bedrooms):
    loc_index = np.where(X_AXIS.columns==location)[0][0]
    
    x=np.zeros(len(X_AXIS.columns))
    x[0]=sqft
    x[1]=bath
    x[2]=bedrooms
    if loc_index>=0:
        x[loc_index] = 1
#Returns numerical value in Lacs        
    return lr_clf.predict([x])[0]
```
