```python
#
#plotting histogram to check no. of properties in price per square ft.
#

import matplotlib
matplotlib.rcParams["figure.figsize"] = (20,10)
plt.hist(df3.ChainOutlet,rwidth=0.8)
plt.xlabel("Chain Outlets")
plt.ylabel("Count")
```

```python
#
# Mapping textual data to numerical values to use in Regrssion Model later
#


df3 = df2.copy()

def toNumbers(df,column):
    mapping = {}
    count = 0
    
    for name in df[column].unique():
        mapping[name]=count
        count = count + 1
        
    temp = df
    temp = temp.replace({column:mapping})[column]
    new_col_name = column+" with numbers" 
    temp = temp.rename(new_col_name)
    df = pd.concat([df,temp],axis='columns')
    return df

df3 = toNumbers(df3,'ChainOutlet')
df3 = toNumbers(df3,'Region')
df3 = toNumbers(df3,'What features will hook a retailer on an App')

df3.head()
```

```python
#
#Defining X-Axis of the graph which will be used by Model
#

X_AXIS = df3.drop(['What features will hook a retailer on an App','Region','ChainOutlet','What features will hook a retailer on an App with numbers'],axis='columns')
X_AXIS.head()
```

```python
#
#Defining Y-Axis of the graph
#

Y_AXIS = df3['What features will hook a retailer on an App with numbers']
Y_AXIS.head()
```

```python
#
#Importing the training and testing model from sklearn
#

from sklearn.model_selection import train_test_split

#
#Setting the training and testing data
#
X_AXIS_train, X_AXIS_test, Y_AXIS_train, Y_AXIS_test = train_test_split(X_AXIS,Y_AXIS,test_size=0.2,random_state=10)
```

```python
#
#Using Linear Regression to fit the curve and predict a score
#We can also use multiple other methods(Decision Tree, Lasso etc..) but Linear Regression is giving best accuracy
#

from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_AXIS_train,Y_AXIS_train)
print("Accuracy : "+str((lr_clf.score(X_AXIS_test,Y_AXIS_test)*100)*100) + "%")
```

```python
#Importing Selection Model
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score



#Randomizes Samples so each fold has equal distribution of data
cv = ShuffleSplit(n_splits=100, test_size=0.9, random_state=3)
print("Randomizing Samples : Done")


#K-Fold Cross Validation
cross_val_score(LinearRegression(),X_AXIS,Y_AXIS,cv=cv)
print("Applying K-Fold Cross Validation : Done")
```

```python
#Function to predict value using Linear Regression
def Predict(region,chain):
    x=np.zeros(len(X_AXIS.columns))
    x[0]=chain
    x[1]=region        
    num_ans = lr_clf.predict([x])[0]
    print(num_ans)
    num_ans = round(num_ans)
    ans = df3.loc[df3['What features will hook a retailer on an App with numbers']==num_ans,'What features will hook a retailer on an App'].iloc[0]
    return ans
```
